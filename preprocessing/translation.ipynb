# Translation from danish to english

## General Translation

```python
df_survey_eng = df_survey_dk.copy()
df_survey_eng.rename(columns={'koen': 'gender', 'Alder': 'age'}, inplace=True)
df_survey_eng['gender'].replace(to_replace=['Mand', 'Kvinde'], value=['Man', 'Woman'], inplace=True)
df_survey_eng['age'] = df_survey_eng['age'].str.replace(' �r', '')
df_survey_eng['region'].replace(to_replace=['Hovedstaden', 'Sj�lland', 'Midtjylland', 'Syddanmark', 'Nordjylland'],
                            value=['Capital', 'Zealand', 'Central', 'Southern', 'Northern'], inplace=True)
df_survey_eng.replace(to_replace=['Nej', 'Ja', 'Ved ikke', 'Ved ikke / �nsker ikke at svare', '�nsker ikke at svare'], value=['no', 'yes', 'dont know', 'dont know/dont want to answer', 'dont want to answer'], inplace=True)
df_survey_eng.head()
```

## Translate Q10 - Q14

```python
df_survey_eng['Q10'].replace(to_replace=['Ph.D og forskeruddanelser', 'Erhvervsfaglig uddannelse (fx t�mrer, fris�r eller kontorassistent)', 
                                   'Korte videreg�ende (fx laborant eller datamatiker)', 'Grundskole (folkeskole, privatskole, efterskole)',
                                   'Lang videreg�ende (fx gymnasiel�rer, �konom eller jurist)', 'Mellemlang videreg�ende (fx teknikum ingeni�r eller l�rer)',
                                   'Gymnasie (STX, HF, HHX , HTX og lign.)','Bachelor (fx 1. del af en lang videreg�ende uddannelse)'],
                       value=['Ph.D', 'Vocational training', 'Short higher education', 'Elementary school', 'Long higher education',
                              'Medium higher education','High school', 'Bachelor'], inplace=True)
df_survey_eng['Q11'].replace(to_replace=['1 person', '2 personer', '4 personer', '3 personer', '6 personer eller derover', '5 personer'], 
                       value=['1 person', '2 people', '4 people', '3 people', '6 people or more', '5 people'], inplace=True)
df_survey_eng['Q12'].replace(to_replace=['Jeg har ingen hjemmeboende b�rn', '1 barn', '2 b�rn', '3 b�rn', '5 b�rn', '4 b�rn', '6 b�rn eller derover'], 
                       value=['I have no children living at home', '1 child', '2 children', '3 children', '5 children', '4 children', '6 children or more'], inplace=True)
df_survey_eng['Q14'].replace(to_replace=['Pensionist eller efterl�nsmodtager', 'L�nmodtager i en privat virksomhed', 'Arbejdsl�s',
                                  'L�nmodtager i det offentlige (stat, kommune eller region)', 'Andet', 'Selvst�ndig', 'Under uddannelse'], 
                       value=['Pensioner or early retirement recipient', 'Employee in a private company', 'Unemployed',
                                  'Employee in the public sector (state, municipality or region)', 'Other', 'Self empoyed', 'Student'], inplace=True)
```

## Translate Q15 & Q16

```python
party_names_danish = ['A: Socialdemokratiet', 'B: Radikale', 'C: Konservative', 'D: Nye Borgerlige', 'E: Klaus Riskj�r Pedersen', 'F: Socialistisk Folkeparti',
                      'I: Liberal Alliance', 'K: Kristendemokraterne', 'O: Dansk Folkeparti', 'P: Stram Kurs', 'V: Venstre', '�: Enhedslisten', '�: Alternativet']
party_names_english = ['A: Social Democrats', 'B: Danish Social Liberal Party', 'C: Conservative Peoples Party', 'D: The New Right', 'E: Klaus Riskjaer Pedersen', 'F: Socialist Peoples Party',
                      'I: Liberal Alliance', 'K: Christian Demokrats', 'O: Danish Peoples Party', 'P: Hard Line', 'V: Venstre', 'OE: Red-Green Alliance', 'Ao: The Alternative']
df_survey_eng.replace(to_replace=party_names_danish, value=party_names_english, inplace=True)

df_survey_eng.replace(to_replace=['Vil ikke svare', 'Havde ikke stemmeret', 'Stemte ikke', 'Stemte blankt', 'Andre partier/kandidat uden for partierne',
                              'Er ikke stemmeberettiget', 'I tvivl om partivalg', 'Vil ikke stemme', 'Vil stemme blankt'],
                         value=['Will not answer', 'Did not have right to vote', 'Did not vote', 'Voted blank', 'Other parties/candidate outside parties',
                                'Not eligible to vote', 'Unsure of party choice', 'Would not vote', 'Would vote blank'], inplace=True)
```
