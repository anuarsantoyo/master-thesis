# Metrics for optimal number of latent factors (clusters)


## [Cophenetic correlation coefficient](https://stats.stackexchange.com/questions/111205/how-to-choose-an-optimal-number-of-latent-factors-in-non-negative-matrix-factori)
 You repeat NMF several time per rank and you calculate how similar are the results. In other words, how stable are the identified clusters, given that the initial seed is random. Choose the highest K before the cophenetic coefficient drops.

```python
 # Caution, takes a while
stability = []
for k in range(2,5):
    H_list = []
    for i in tqdm(range(5), desc=f"k:{k} original"):
        W, H, n_iter = non_negative_factorization(answers, n_components=k,max_iter=200,init='random')
        H_list.append(H)
    Hs_matrix = np.stack(H_list)
    H_mean = Hs_matrix.mean(axis=0)
    Hs_matrix_diff = Hs_matrix - H_mean
    stability.append(np.square(Hs_matrix_diff).mean())

plt.plot(stability,label="mse")
plt.title("NMF Elbow Method (largest k before drop)")
plt.xlabel('K')
plt.ylabel('mse')
plt.legend()
plt.show()
```

## [Residual sum of squares against randomized data (RSS)](https://stats.stackexchange.com/questions/111205/how-to-choose-an-optimal-number-of-latent-factors-in-non-negative-matrix-factori)
For any dimensionality reduction approach, there is always a loss of information compared to your original data (estimated by RSS). Now perform NMF for increasing K and calculate RSS with both your original dataset and a randomized dataset. When comparing RSS in function of K, the RSS decreases with increasing K in the original dataset, but this is less the case for the randomized dataset. By comparing both slopes, there should be an K where they cross. In other words, how much information could you afford to lose (=highest K) before being within the noise.

```python
np.random.seed(0)
answers_rand = (np.random.rand(*answers.shape)*7).round() # Create random answer matrix
```

```python
# Caution, takes a lot of time to run
mse_original = []
mse_random = []
for k in tqdm(range(2,10)):
    W, H, n_iter = non_negative_factorization(answers, n_components=k,max_iter=200,init='random')
    mse_original.append(np.square(answers - np.matmul(W,H)).mean())

    W, H, n_iter = non_negative_factorization(answers_rand, n_components=k,max_iter=200,init='random')
    mse_random.append(np.square(answers_rand - np.matmul(W,H)).mean())
```

```python
plt.plot(mse_original,label="original")
plt.plot(mse_random, label="random")
plt.xlabel('k-2')
plt.ylabel('mse')
plt.title("Error |Answers - W*H|²")
plt.legend()
plt.show()

orig_diff = np.diff(mse_original)
rand_diff = np.diff(mse_random)
plt.plot(orig_diff,label="original")
plt.plot(rand_diff, label="random")
plt.xlabel('k-2')
plt.ylabel('mse')
plt.title("Slope of Error")
plt.legend()
plt.show()
```
